```
# Enter your code here. Read input from STDIN. Print output to STDOUT
i = input()
i = i.split()
n = int(i[0])
m = int(3 * n)
if n != 5:
    x = int(n - 5)
else:
    x = 0
if n != 5:
    y = int(x/2)
else:
    y = 0
base = '''------.|.------
---.|..|..|.---
----WELCOME----
---.|..|..|.---
------.|.------'''
    
new_base = base.replace("----WELCOME----", 3 * y * '-' + '----WELCOME----' + 3 * y * '-')
new_base = new_base.replace('---.|..|..|.---', '---.|..' + y * '|..' + '|' + y * '..|' + '..|.---')
new_base = new_base.replace('------.|.------', '------.' + y * '|..' + '|' + y * '..|' + '.------')
# new_line = y * 3 * '-' + '------.|.------' + y * 3 * '-'

for i in range(1, y+1):
    new_line = i * '---' + '------.' + (y - i) * '|..' + '|' + (y - i) * '..|' + '.------' + i * '---' 
    new_base =  new_line + '\n' + new_base + '\n' + new_line 
    
print(new_base)
```
